```python
import os

import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd

datafile = os.path.expanduser('/home/dave/projects/diploma/data/place.csv')
datafile2 = os.path.expanduser('/home/dave/projects/diploma/maps/data/API_IT.NET.USER.ZS_DS2_en_csv_v2_10082442.csv')
nn_file = os.path.expanduser('/home/dave/projects/diploma/maps/nn_output/something2.csv')
shapefile = os.path.expanduser('/home/dave/projects/diploma/maps/data/ne_10m_admin_0_countries_lakes.shp')

cols = ['id', 'code', "name"]

colors = 9
cmap = 'Blues'
figsize = (16, 10)
year = '2016'
cols2 = ['Country Name', 'Country Code']
title = 'Individuals using the Internet (% of population) in {}'.format(year)
imgfile = 'img/{}.png'.format((title)) #here was slug

description = '''
Individuals who have used the Internet from any location in the last 3 months via any device based on the International Telecommunication Union,
World Telecommunication/ICT Development Report and database. Data: World Bank - worldbank.org • Author: Ramiro Gómez - ramiro.org'''.strip()
```

```python
gdf = gpd.read_file(shapefile)[['ADM0_A3', 'geometry']].to_crs('+proj=robin')
gdf.sample(5)
```

```python
df = pd.read_csv(datafile, skiprows=0, usecols=cols, delimiter=';')
df.sample(5)
```

```python
df2 = pd.read_csv(datafile2, skiprows=4, usecols=cols2)
df2.sample(5)


```

```python
nn = pd.read_csv(nn_file, skiprows=0, delimiter=';')
nn.sample(5)
```

```python
merged = nn.merge(df, left_on='id', right_on='id')
merged.sample(5)
```

```python
merged2 = merged.merge(df2, left_on='name', right_on='Country Name')
merged2.sample(5)
```

```python
merged3 = gdf.merge(merged2, left_on='ADM0_A3', right_on='Country Code')
merged3.describe()
```

```python
prediction = "prediction"
ax = merged3.dropna().plot(column=prediction, cmap=cmap, figsize=figsize, scheme='equal_interval', k=colors, legend=True)
```
