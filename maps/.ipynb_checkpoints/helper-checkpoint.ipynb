```python
%matplotlib inline

import os

import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd

places_df = os.path.expanduser('/home/dave/projects/diploma/data/place_code.csv')
API_df = os.path.expanduser('/home/dave/projects/diploma/maps/data/API_IT.NET.USER.ZS_DS2_en_csv_v2_10082442.csv')
nn_file = os.path.expanduser('/home/dave/projects/diploma/maps/nn_output/something2.csv')
shapefile = os.path.expanduser('/home/dave/projects/diploma/maps/data/ne_10m_admin_0_countries_lakes.shp')

cols = ['id', 'code', "name",'country code']

colors = 9
cmap = 'Blues'
figsize = (16, 10)
year = '2016'
cols2 = ['Country Name', 'Country Code', '2016']
title = 'Individuals using the Internet (% of population) in {}'.format(year)
imgfile = 'img/{}.png'.format((title)) #here was slug

description = '''
Individuals who have used the Internet from any location in the last 3 months via any device based on the International Telecommunication Union,
World Telecommunication/ICT Development Report and database. Data: World Bank - worldbank.org • Author: Ramiro Gómez - ramiro.org'''.strip()
```

```python
shapes_pd = gpd.read_file(shapefile)[['ADM0_A3', 'geometry']].to_crs('+proj=robin')
shapes_pd.sample()
```

```python
places_pd = pd.read_csv(places_df, skiprows=0, usecols=cols, delimiter=';')
places_pd.sample(5)
```

```python
api_pd = pd.read_csv(API_df, skiprows=4, usecols=cols2)
api_pd.count()
```

```python
nn_pd = pd.read_csv(nn_file, skiprows=0, delimiter=';')
nn_pd.sample(5)
```

```python
places_nn_merge = places_pd.merge(nn_pd, left_on='id', right_on='id')
places_nn_merge.sample(5)
#merged.describe()
```

```python
country_code_merge = api_pd.merge(places_nn_merge, right_on='country name', left_on='Country Name', how='outer')
country_code_merge.count()
#merged2.describe()
```

```python
shape_merge = shapes_pd.merge(country_code_merge, left_on='ADM0_A3', right_on='Country Code', how='inner')
shape_merge.drop(columns=['id', 'code', 'name'])
shape_merge.count()
```

```python
merged = shapes_pd.merge(api_pd, left_on='ADM0_A3', right_on='Country Code')
merged.sample(5)
```

```python
ax = shape_merge.dropna().plot(column='prediction', cmap=cmap, figsize=figsize, scheme='equal_interval', k=colors, legend=True)
ax.get_figure().savefig('map_export.png',dpi=300)
```
